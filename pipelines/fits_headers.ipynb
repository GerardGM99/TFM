```python
import os
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
import astropy.units as u
import numpy as np
```

```python
# Path to the directory containing FITS files
directory = '20240412'

# Initialize an empty list to store header information
header_info_list = []

# Loop through each FITS file in the directory
for filename in os.listdir(directory):
    if filename.endswith('0021.fits'):
        filepath = os.path.join(directory, filename)
        
        # Open the FITS file
        with fits.open(filepath) as hdul:
            # Extract header information from the primary header
            header = hdul[0].header
            
            # Extract important information from the header
            # Example: Extracting OBJECT, EXPTIME, DATE-OBS
            naxis1 = header.get('NAXIS1', 'N/A')
            naxis2 = header.get('NAXIS2', 'N/A')
            airmass = header.get('AIRMASS', 'N/A')
            utc = header.get('UTC', 'N/A')
            
            
            # Append the information to the list
            header_info_list.append({'Filename': filename,
                                     )

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(header_info_list)

# Display the DataFrame
print(df)
```

```python
# Path to the directory containing FITS files
directory = '20240412'

# Initialize an empty list to store header information
header_info_list = []

# Loop through each FITS file in the directory
for i in range(41, 48):  # Files from uniB_0021.fits to uniB_0040.fits
    filename = f'uniB_{i:04d}.fits'
    filepath = os.path.join(directory, filename)
    
    # Open the FITS file
    with fits.open(filepath, mode='update') as hdul:
        # Extract and modify the desired header
        header = hdul[0].header
        header['IMAGETYP'] = 'arc'  # Change a specific header value
        
        # Save the changes (optional)
        hdul.flush()
```

```python
input_filename = '20240412/spectra1D_dswfz_uniB_0118.fits'

data = fits.getdata(input_filename)
header = fits.getheader(input_filename)

naxis2, naxis1 = data.shape
crpix1 = header['crpix1'] * u.pixel
cunit1 = 1 * u.Unit(header['cunit1'])
crval1 = header['crval1'] * u.Unit(cunit1)
cdelt1 = header['cdelt1'] * u.Unit(cunit1) / u.pixel
wavelength = crval1 + ((np.arange(naxis1) + 1)*u.pixel - crpix1) * cdelt1

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(wavelength.to(u.Angstrom), data[0])
ax.set_xlabel('Wavelength (Angstrom)')
ax.set_ylabel('Flux (number of counts)')
plt.title (input_filename)
plt.show()
```
